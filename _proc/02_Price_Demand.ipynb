## Price Elasticity

>Price elasticity of demand is the ratio of the percentage change in quantity demanded of a product to the percentage change in price.

<center>$Price Elasticity = (d(Quantity)/Quantity)) / (d(Price)/Price))$</center>

## Regression Overview
Regression analysis is a set of statistical methods used for the estimation of relationships between a dependent variable and one or more independent variables. It can be utilized to assess the strength of the relationship between variables and for modeling the future relationship between them.

#### Equation:
> <center style="font-size: 30px">$Y = a + b*x1 + c*x2 + d*x3 + ... + e$</center>
Where:<br>
Y – Dependent (target) variable<br>  
x1, x2, x3 – Independent (explanatory) variables<br>
a – Intercept<br>
b, c, d – Slopes<br>
e – Residual (error)<br>

Interpretation: one unit change in variable x1 will impact Y by the factor of b, keeping all the values as it is
