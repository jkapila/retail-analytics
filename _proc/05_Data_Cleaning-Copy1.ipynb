```python
# Import Libraries
from nbdev.showdoc import *
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
import re

# Load Dataset
df=pd.read_csv('Finance Data/train.csv')
```

Delete unwanted columns

```python
del df['ID'] # Identification 
del df['SSN'] # SSN (social security number of a person)
```

```python
df
```

Remove everything except numbers

```python
num_cols=['Age', 'Annual_Income', 'Num_of_Loan', 'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 'Amount_invested_monthly', 'Outstanding_Debt' , 'Monthly_Balance']

def keep_num(val):
    try : 
        return float(re.sub("[^-\d.]","",val))
    except :
        return np.nan
    
for col in num_cols:
    df[col]=df[col].apply(keep_num)
```

```python
df.head()
```

Split loan type column

```python
loan_t=df['Type_of_Loan'].drop_duplicates().str.split(',',expand=True)
loan_type_ls=loan_t[0].drop_duplicates().dropna()
for i in loan_type_ls: 
    df[i] = df['Type_of_Loan'].str.contains(i)
```

```python
# filter negative age
df=df[(df['Age']>18) & (df['Age']<120)]
# filter negative num of bank account
df=df[df['Num_Bank_Accounts']>0]
# filter negative num of loan
df=df[df['Num_of_Loan']>0]
# filter negative num of delayed payment
df=df[df['Num_of_Delayed_Payment']>0]
# filter negative monthly balance
df=df[df['Monthly_Balance']>0]
# filter 0 or negative number of credit cards
df=df[df['Num_Credit_Card']>0]

df
```

```python
# credit history age from yr mon to months
def extract_num(string):
    return int(re.sub('[^0-9,]', "", string))

def to_months(Yr_mon):
    try:
        return sum([extract_num(s)*12 if 'Year' in s else extract_num(s) for s in Yr_mon.split('and')])
    except:
        return np.nan
    
df['Credit_History_Age']=df.Credit_History_Age.apply(to_months)
```

```python
# remove special character from payment behaviour
def clean_pay_behaviour(string):
    try:
        return re.sub('[^a-zA-Z_]', "", string)
    except:
        return np.nan
df.Payment_Behaviour=df.Payment_Behaviour.apply(clean_pay_behaviour)
```

```python
df.Month=df.Month.apply(lambda s: datetime.strptime(s, '%B').month)
```

```python
df.Occupation=df.Occupation.replace('_______',np.nan)
df.Credit_Mix=df.Credit_Mix.replace('_',np.nan)
```

```python
f = lambda x: x.median() if np.issubdtype(x.dtype, np.number) else next(iter(x.mode()), np.nan) 
# df = df.fillna(df.groupby('make').transform(f))
```

```python
df=df.fillna(df.groupby('Customer_ID').transform(f))
```
