Data contains historical sales (`2015` to `2018`) of products in supermart. These products are categorized in `7` category:<br>

### Data Dictionary

>`Order ID`: *Unique Id assign to each order* <br>
`Customer Name`: *Name of the customer who purchased*<br>
`Customer Id`: *Unique Id assign to each customer*<br>
`Age`: *Age of the customer*<br>
`Gender`: *Gender type (ex. Male and Female)*<br>
`Category`: *Category of the product (ex. Bakery, Beverages, etc.)*<br>
`Sub Category`: *Sub Cateogry of the product (ex. Biscuits, Cakes, etc.)*<br>
`Region`: *Central, East, North, South, West*<br>
`State`: *State where products are purchased*<br>
`City`: *City where products are purchased*<br>
`Order Date`: *Purchase date*<br>
`Time`: *Purchase time*<br>
`Payment Method`: *Payment used by customer for purchase (ex. Cash, Ewallet, Credit card)*<br>
`Sales`: *Total sales amount *<br>
`Quantity`: *Quantity purchased*<br>
`Discount`: *Discount percentage*<br>
`Profit`: *Profit amount*<br>
`Rating`: *Customer rating (1 to 10)*<br>
